```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("HN-ratio_human_221027_All.csv", sep=",")
df.head()
```

```python
# set index 
df1=df.set_index('GeneName')
df1.head()
```

```python

temp = pd.DataFrame()
thre = df1>=1.5
temp["up1.5"] = thre.sum(axis=1)

thre = df1>=2
temp["up2"] = thre.sum(axis=1)

thre = df1>=5
temp["up5"] = thre.sum(axis=1)

thre = df1>=10
temp["up10"] = thre.sum(axis=1)

thre = df1<=1/1.5
temp["dn1.5"] = thre.sum(axis=1)

thre = df1<=1/2
temp["dn2"] = thre.sum(axis=1)

thre = df1<=1/5
temp["dn5"] = thre.sum(axis=1)

thre = df1<=1/10
temp["dn10"] = thre.sum(axis=1)

thre = df1 >= 0
temp["all"] = thre.sum(axis=1)

temp["unchange1.5"] = temp["all"] - temp["up1.5"] - temp["dn1.5"]
temp["unchange2"] = temp["all"] - temp["up2"] - temp["dn2"]
temp["unchange5"] = temp["all"] - temp["up5"] - temp["dn5"]
temp["unchange10"] = temp["all"] - temp["up10"] - temp["dn10"]

temp["HN1.5"] = temp["up1.5"] - temp["dn1.5"]
temp["HN2"] = temp["up2"] - temp["dn2"]
temp["HN5"] = temp["up5"] - temp["dn5"]
temp["HN10"] = temp["up10"] - temp["dn10"]


temp
```

```python
now = datetime.datetime.now()
print(now)
```

```python
temp=temp.reset_index()
```

```python
tempHN2= temp.sort_values(by= "HN2", ascending=False)
tempHN2
```

```python
now = datetime.datetime.now()

tempHN2.to_csv('./HN-score_human_' + now.strftime('%y%m%d') + '_HN2' + '_all.csv', sep=",")
```
