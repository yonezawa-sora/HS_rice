```python
import pandas as pd
import numpy as np
import datetime
import os
```

```python
df = pd.read_csv("./HN-ratio_rice_231005_All.csv", sep=",")
df.head()
```

```python
# set index 
df1=df.set_index('GENEID')
df1.head()
```

```python
temp = pd.DataFrame()

# Upregulated
thresholds = [1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10, 10.5, 11, 12, 15, 20, 50, 75,100, 200]  # input thresholds here

for thre in thresholds:
    thre_df = df1 >= thre
    temp[f"up{thre}"] = thre_df.sum(axis=1)

# Downregulated
for thre in thresholds:
    thre_df = df1 <= 1/thre
    temp[f"dn{thre}"] = thre_df.sum(axis=1)

# Unchanged
for thre in thresholds:
    temp[f"unchange{thre}"] = temp[f"up{thre}"] - temp[f"dn{thre}"]

# HN-score
for thre in thresholds:
    temp[f"HN{thre}"] = temp[f"up{thre}"] - temp[f"dn{thre}"]

temp
```

```python
now = datetime.datetime.now()
print(now)
```

```python
# Extract columns with "HN" from tmp
tempHN = temp.filter(like='HN', axis=1)
tempHN
```

```python
# example of sorting by HN5
tempHN5= tempHN.sort_values(by= "HN5", ascending=False)
tempHN5.to_csv("HN-score_rice_HN5_example_" + now.strftime('%y%m%d') + ".csv")
tempHN5
```

```python
now = datetime.datetime.now()

# making directory each time
directory = f'./HN-score_rice_{now.strftime("%y%m%d")}'
os.makedirs(directory, exist_ok=True) # exist_ok=True: if the directory already exists, no error

# Using f-strings for formatting strings (e.g. temp[f"HN{thre}"])
for thre in thresholds:
    sorted_tempHN = tempHN.sort_values(by=f"HN{thre}", ascending=False) # sort by HN-score
    sorted_tempHN[f"HN{thre}"].to_csv(f'./{directory}/HN-score_rice_{now.strftime("%y%m%d")}_HN{thre}_all.csv', sep=",", index=True) # save as csv
```
