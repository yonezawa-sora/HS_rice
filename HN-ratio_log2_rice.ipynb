```python
# import library for expression-ratio analysis
import pandas as pd
import numpy as np
import datetime
```

```python
now = datetime.datetime.now()
print(now)
```

```python
# 全ての発現データを載せたファイルをアップロード
df = pd.read_csv(filepath_or_buffer="./")
df.head()
```

```python
# scaled TPMなので足すと全て同じ数値になることを示している
df.describe()
```

```python
HN = pd.read_csv('HN-pair-human_All.csv' )
HN
```

```python
HNratio = pd.DataFrame()
HNratio["GeneName"] = df["GeneName"] 
for i in range(len(HN)):
    h = str(HN.iloc[i,0])
    n = str(HN.iloc[i,1])
#    c = str(h) +"/"+str(n)
    HNratio[h] = (df.loc[:,h] + 1)/(df.loc[:,n] +1)
    HNratio.append([HNratio, HNratio[h]])
HNratio.head()
```

```python
HNratio_ = HNratio.set_index('GeneName')

HNratiolog2 = HNratio_.apply(np.log2)
HNratiolog2 = HNratiolog2.reset_index()
HNratiolog2.head()
```

```python
HNratiolog2.describe()
```

```python
HNratiolog2.set_index('GeneName').to_csv("HN-ratio_log2_human_ " + now.strftime('%y%m%d') + "_All" + ".csv", sep=",")
```
