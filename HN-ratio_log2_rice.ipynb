```python
# import library
import pandas as pd
import numpy as np
import datetime 
```

```python
now = datetime.datetime.now()
print(now)
```

```python
#Upload gene expression data (quantitated by salmon v1.10)
df = pd.read_table(filepath_or_buffer="./rice_230910.tsv", sep="\t")
df.head()
```

```python
#This indicates that the count data will be the same for all samples 
df.describe()
```

```python
HN = pd.read_csv("./HN-pair_rice.csv", sep=",")
HN
```

```python
HNratio = pd.DataFrame()
HNratio["GENEID"] = df["GENEID"] 
for i in range(len(HN)):
    h = str(HN.iloc[i,0])
    n = str(HN.iloc[i,1])
#    c = str(h) +"/"+str(n)
    HNratio[h] = (df.loc[:,h] + 1)/(df.loc[:,n] +1)
    HNratio.append([HNratio, HNratio[h]])
HNratio.head()
```

```python
HNratio_ = HNratio.set_index('GENEID')

HNratiolog2 = HNratio_.apply(np.log2)
HNratiolog2 = HNratiolog2.reset_index()
HNratiolog2.head()
```

```python
HNratiolog2.describe()
```

```python
HNratiolog2.set_index('GENEID').to_csv("HN-ratio_log2_rice_ " + now.strftime('%y%m%d') + "_All" + ".csv", sep=",")
```
